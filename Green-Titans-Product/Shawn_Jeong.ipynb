```python
# Basic Imports
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib as mb
import matplotlib.pyplot as plt
```

```python
# classification: binary because having a stroke is a true or false
```

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
# ^ The pie chart above shows the the smoking status of the diffrent people in the data set. This can show the diffrent distrobutions of diffrent smoking status among the sam
```

```python
# The scatter plot above is explaining how with the increasing age, the BMI values tend to stay around the same, but the average glucose level tends increase with age
```

```python
def clean_data():
    global df
    df = df[df['bmi'].notna()]
    columns_to_drop = ['id', 'ever_married', 'work_type']
    df.drop(columns_to_drop, axis=1, inplace=True)
    df.drop(df.index[df['gender'] == 'Other'], inplace = True)

def pie():
    global df
    new_df = df.groupby('smoking_status').count()
    options = list(new_df.index)
    counts = []
    for val in options:
        counts.append(new_df["heart_disease"][val])

    new_fig = px.pie(new_df, values=counts, names=options, title="Percentage of Smokers")
    new_fig.show()

def scat():
    global df
    new_fig = px.scatter(df, x="age", y="bmi",color="avg_glucose_level")
    new_fig.show()

def smoke_hist():
    global df
    fig = px.histogram(df, x="smoking_status", y="avg_glucose_level", color='gender', barmode='group',height=400)
    fig.show()
```

```python
clean_data()
```
