```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
```

```python
sd = pd.read_csv("healthcare-dataset-stroke-data.csv")
#hypertension = high blood pressure
sd.dropna(axis = 0, how = 'any', inplace = True)
sd.drop(sd.index[sd['gender'] == 'Other'], inplace = True)

sd1 = sd.loc[sd['avg_glucose_level'] <= 168.32]
sd = sd1
sd.reset_index(drop = True, inplace = True)

sd2 = sd.loc[sd['bmi'] <= 47.5]
sd = sd2
sd.reset_index(drop = True, inplace = True)

sd.drop('id',inplace=True,axis=1)

sd.head(10)
```

```python
genderdummy = pd.get_dummies(sd["gender"])
sd["gender"] = genderdummy["Male"]
marrydummy = pd.get_dummies(sd["ever_married"])
sd["ever_married"] = marrydummy["Yes"]
reddummy = pd.get_dummies(sd["Residence_type"])
sd["Residence_type"] = reddummy["Urban"]
smokedummy = pd.get_dummies(sd["smoking_status"])
sd["smoking_status"] = smokedummy["never smoked"]
sd["smoking_status"] = sd["smoking_status"].replace([0],2)
sd["smoking_status"] = sd["smoking_status"].replace([1],0)
sd["smoking_status"] = sd["smoking_status"].replace([2],1)
```

```python
sd.head(10)
```

## EDA - Exploratory Data Analysis



```python
def pie():
    global sd
    new_stroke_data = sd.groupby('smoking_status').count()
    options = list(new_stroke_data.index)
    counts = []
    for val in options:
        counts.append(new_stroke_data["heart_disease"][val])

    new_fig = px.pie(new_stroke_data, values=counts, names=options, title="Percentage of Smokers")
    new_fig.show()
def scat():
    global sd
    new_fig = px.scatter(sd, x="age", y="bmi",color="avg_glucose_level")
    new_fig.show()

def smoke_hist():
    global sd
    fig = px.histogram(sd, x="smoking_status", y="avg_glucose_level", color='gender', barmode='group',height=400)
    fig.show()
```

```python
pie()
```

```python
smoke_hist()
```

```python
fig = px.scatter_matrix(sd,
    dimensions=["age", "avg_glucose_level", "bmi"],
    color="stroke")
fig.show()
```

The older you are, the more likely you are to have a stroke. The higher you average glucose level is, the higher the stroke risk becomes. The higher your BMI is, the more likely you ate to have a stroke

The most dangerous combination is a high glucose level, and old age

The only data point that does not have any direct relation to strokes is age, although it does raise the chance for a person to get a stroke, when they are older
